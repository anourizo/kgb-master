```python
import numpy as np
from itertools import chain

# 1) PARAMETERS FOR YOUR THREE SIMULATIONS
sim1 = dict(path='/home/ahmad/kgb-master/data/3072-90000/',
            ngrid=3072, boxsize=90000)
sim2 = dict(path='/home/ahmad/kgb-master/data/4096-15000/',
            ngrid=4096, boxsize=15000)
sim3 = dict(path='/home/ahmad/kgb-master/data/3072-4000/',
            ngrid=3072, boxsize=4000)

# List of all the P(k) “suffixes” you want to load:
pknames = [
    'delta_kgb',
    'delta',
    'phi',
    'phi_prime',
    'pi_k',
    'zeta',
    'deltakgb_deltam',
]

# 2) READ YOUR z-list once (from sim1’s settings ini)
z = []
with open(f"{sim1['path']}/file_settings_used.ini") as f:
    for line in f:
        if line.strip().startswith('Pk redshifts'):
            z = [float(x) for x in line.split('=')[1].split('#')[0].split(',')]
            break

def load_simulation(path, z, pknames):
    """Load all pk_0NN_<name>.dat files into a dict of lists."""
    data = {name: [] for name in ['k'] + pknames}
    for i in range(len(z)):
        idx = f"{i:02d}"
        first = True
        for name in pknames:
            arr = np.loadtxt(f"{path}/pk_0{idx}_{name}.dat")
            if first:
                data['k'].append(arr[:,0])
                first = False
            data[name].append(arr[:,1])
    return data

def trim_snapshot(k, fields, ngrid, boxsize,
                  remove_first=3, nyquist_frac=1/13):
    """
    Remove the first `remove_first` modes and all modes above
    `nyquist_frac * (pi*ngrid/boxsize)`.
    """
    kny = np.pi * ngrid / boxsize
    cutoff = nyquist_frac * kny
    iend = np.argmax(k > cutoff)
    keep = slice(remove_first, iend)
    return k[keep], {name: arr[keep] for name, arr in fields.items()}

# --- load all three sims ---
data1 = load_simulation(sim1['path'], z, pknames)
data2 = load_simulation(sim2['path'], z, pknames)
data3 = load_simulation(sim3['path'], z, pknames)

# 3) stitch them together
final = {name: [] for name in ['k'] + pknames}

for iz in range(len(z)):
    # trim sim1 & sim2, stitch into inter_k, inter_f
    k1, f1 = trim_snapshot(data1['k'][iz],
                           {n: data1[n][iz] for n in pknames},
                           sim1['ngrid'], sim1['boxsize'])
#     k2, f2 = trim_snapshot(data2['k'][iz],
#                            {n: data2[n][iz] for n in pknames},
#                            sim2['ngrid'], sim2['boxsize'])

    k2, f2 = trim_snapshot(
        data2['k'][iz],
        {n: data2[n][iz] for n in pknames},
        sim2['ngrid'], sim2['boxsize'],
        remove_first=3,
        nyquist_frac=1/10
    )
    j2 = np.searchsorted(k2, k1[-1], side='right')
    inter_k = np.concatenate([k1, k2[j2:]])
    inter_f = {name: np.concatenate([f1[name], f2[name][j2:]]) for name in pknames}

#     # now trim sim3 and append its high-k end
#     k3, f3 = trim_snapshot(data3['k'][iz],
#                            {n: data3[n][iz] for n in pknames},
#                            sim3['ngrid'], sim3['boxsize'])
#     j3 = np.searchsorted(k3, inter_k[-1], side='right')

    k3, f3 = trim_snapshot(
        data3['k'][iz],
        {n: data3[n][iz] for n in pknames},
        sim3['ngrid'], sim3['boxsize'],
        remove_first=23,
        nyquist_frac=0.5
    )
    j3 = np.searchsorted(k3, inter_k[-1], side='right')

    # final concatenation
    final['k'].append(np.concatenate([inter_k, k3[j3:]]))
    for name in pknames:
        final[name].append(
            np.concatenate([inter_f[name], f3[name][j3:]])
        )

# 4) normalize only the 'delta' spectrum by (2π² / k³)
for iz in range(len(z)):
    ktot = final['k'][iz]
    norm = (2 * np.pi**2) / (ktot**3)
    final['delta'][iz] *= norm

# Now final['k'][iz], final['delta'][iz], etc. include all three simulations

```
