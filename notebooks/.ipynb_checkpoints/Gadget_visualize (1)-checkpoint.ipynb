```python
import pyvista as pv
import numpy as np

# Parameters
boxsize = 1024.0  # Mpc/h
ngrid = 256  # Must match your velocity data dimensions
bulk_velocity = np.stack([data[:,:,:,0], data[:,:,:,1], data[:,:,:,2]], axis=-1) * 1.e9
# Assuming bulk_velocity is shape (ngrid, ngrid, ngrid, 3)
# Create coordinate grid (physical space)
x = y = z = np.linspace(0, boxsize, ngrid)
X, Y, Z = np.meshgrid(x, y, z, indexing='ij')

# Create structured grid and add vector field
grid = pv.StructuredGrid(X, Y, Z)
grid['vectors'] = bulk_velocity.reshape(-1, 3)  # Flatten to (ngrid^3, 3)

# Generate seed points throughout entire volume
n_seeds = 10000  # Total number of seeds
np.random.seed(42)  # For reproducibility
seed_points = np.random.rand(n_seeds, 3) * boxsize

# Create seed source
seed_source = pv.PolyData(seed_points)

# Generate streamlines
streamlines = grid.streamlines_from_source(
    seed_source,
    vectors='vectors',
    integration_direction='both',
    max_time=1700.0,           # Max parametric length (~2/3 of box)
    initial_step_length=15.0,
    max_step_length=50.0,
    terminal_speed=1.
)

# Create plotter
plotter = pv.Plotter(window_size=(1000, 1000))
tube_radius = boxsize / ngrid / 5  # Adjust tube thickness

# Add color-mapped streamlines
streamlines['speed'] = np.linalg.norm(streamlines['vectors'], axis=1)
speed = streamlines['speed']
log_speed = np.log10(np.clip(speed, a_min=1., a_max=None))  # or choose a suitable min
streamlines['log_speed'] = log_speed

plotter.add_mesh(
    streamlines.tube(radius=tube_radius),
    scalars='log_speed',
    cmap='Oranges',
    show_scalar_bar=False,
    clim=[np.log10(1.2), np.log10(200)],
    opacity=0.2,  # Slightly more opaque for better visibility
    smooth_shading=True,  # Smooth shading for a more elegant look
    specular=0.5,  # Add some specular lighting
    specular_power=30,  # Increase specular power for sharper highlights
    ambient=0.6  # Increase ambient light
    # show_edges=True,  # Show edges for better structure visibility
    # edge_color='green',  # Edge color
    # silhouette={'color': 'green', 'line_width': 3}  # Add a silhouette

)

# plotter.add_mesh(
#     streamlines.tube(radius=tube_radius),
#     color='orange',
#     opacity=0.35
# )


#### Particles positions weighted with mass:
print("Plotting the points")
particle_positions = pos[::1]
particle_radii = radius_halos[::1]

# Create PyVista object
points = pv.PolyData(particle_positions)

# Add radius data (for coloring if needed)
points['radius'] = particle_radii

# Calculate scaled sizes - ensure they're reasonable
base_size = 1.0
scaling_factor = 3.0
point_sizes = base_size + particle_radii * scaling_factor

# Plot with correct sizing approach
# plotter.add_mesh(
#     points,
#     scalars='radius',  # Optional: color by radius
#     point_size=point_sizes.mean(),  # Base size
#     render_points_as_spheres=True,
#     style='points_gaussian',  # Enables per-point sizing
#     opacity=0.8,
#     cmap='Blues'  # Optional colormap
# )


# Create ultra-high resolution sphere template
sphere = pv.Sphere(
    radius=1.0, 
    theta_resolution=64,
    phi_resolution=64
)

# Create glyphs
glyphs = points.glyph(
    geom=sphere,
    scale='radius',
    factor=1.75,
    orient=False
)

# Advanced rendering with PBR
plotter.add_mesh(
    glyphs,
    color='midnightblue',
    opacity=0.65,
    smooth_shading=True,
    pbr=True,               # Physically Based Rendering
    metallic=0.3,           # 0=non-metal, 1=metal
    roughness=0.2,          # 0=smooth, 1=rough
    diffuse=1.0,
    specular=0.8,
    ambient=0.3
)

# Correct lighting setup (remove problematic environment texture)
plotter.add_light(pv.Light(
    position=(1000, 1000, 1000),
    intensity=0.6,
    light_type='scene light'
))
plotter.add_light(pv.Light(
    position=(-1000, -1000, 1000),
    intensity=0.5,
    light_type='scene light'
))

# Optional: Add depth effects
plotter.enable_anti_aliasing('ssaa')  # Super-sample anti-aliasing

print("Finished plotting the points")

# Configure visualization
# plotter.show_bounds(
#     grid=False,
#     location='outer',
#     all_edges=True,
#     color='black'
# )

plotter.show_bounds(
    grid=False,
    ticks=None,
    xlabel='',
    ylabel='',
    zlabel='',
    xtitle='',
    ytitle='',
    ztitle='',
    show_xaxis=False,
    show_yaxis=False,
    show_zaxis=False,
    show_xlabels=False,
    show_ylabels=False,
    show_zlabels=False,
    location='outer',
    all_edges=True,
    color='grey'
)

original_position = plotter.camera.position
original_focal = plotter.camera.focal_point
original_up = (0.0, 0, 1.)

# Modified camera settings (tilted toward YZ plane)
tilt_factor = 0.4  # Adjust this between 0-1 for more/less tilt (0.2 = 20% shift)

new_position = (
    original_position[0] * (1 - tilt_factor),  # Reduce X component
    original_position[1] * (1 + tilt_factor/1.3),  # Slight Y increase
    original_position[2] * (1 - tilt_factor/1.4)   # Slight Z increase
)

plotter.camera.position = new_position
plotter.camera.focal_point = original_focal
plotter.camera.up = original_up

# Optional: Adjust clipping range if needed
plotter.camera.clipping_range = (1000, 50000)

plotter.set_background("white")
plot_path = "sims_L_1024_N_256.png"
plotter.screenshot(plot_path)
print(f"Saved visualization to: {plot_path}")
```

```python
import numpy as np
import pyvista as pv

# Reduce grid resolution significantly
boxsize = 1024.0  # Mpc/h
ngrid = 32  # Reduced from 64

# Simplified velocity field generation
Vx = np.random.rand(ngrid, ngrid, ngrid)
Vy = np.random.rand(ngrid, ngrid, ngrid)
Vz = np.random.rand(ngrid, ngrid, ngrid)

# Simplified velocity calculation
bulk_velocity = np.stack([Vx, Vy, Vz], axis=-1)

# Construct the grid with fewer points
x = y = z = np.linspace(0, boxsize, ngrid)
X, Y, Z = np.meshgrid(x, y, z, indexing='ij')
grid = pv.StructuredGrid(X, Y, Z)
grid['vectors'] = bulk_velocity.reshape(-1, 3)

# Fewer seed points
seed_ngrid = 8  # Reduced from 16
seed_coords = np.linspace(0, boxsize, seed_ngrid)
SX, SY, SZ = np.meshgrid(seed_coords, seed_coords, seed_coords, indexing='ij')
seeds = np.column_stack([SX.ravel(), SY.ravel(), SZ.ravel()])

# Create source points dataset
source = pv.PolyData(seeds)

# Streamlines with reduced computational intensity
streamlines = grid.streamlines_from_source(
    source,
    vectors='vectors',
    integration_direction='both',
    initial_step_length=boxsize / ngrid,  # Larger step
    terminal_speed=1.0,  # Less strict termination
    max_steps=1000  # Fewer steps
)

# Simplified plotting
plotter = pv.Plotter(window_size=(800, 800))
streamlines['speed'] = np.linalg.norm(streamlines['vectors'], axis=1)
plotter.add_mesh(streamlines.tube(radius=boxsize / ngrid / 10),
                 scalars='speed', cmap='viridis')

plotter.show_bounds(grid=False, location='outer', all_edges=True)
plotter.set_background("white")

# Option 1: Show plot interactively
plotter.show()

# Option 2: Save quickly
# plot_path = "quick_streamlines.png"
# plotter.screenshot(plot_path)
# print(f"Saved visualization to: {plot_path}")
```

```python
import numpy as np
import pyvista as pv

# Parameters
N_particles = 5000
N_grid = 16
box_size = 1.0

# Random clustered particles
np.random.seed(42)
particle_positions = np.random.rand(N_particles, 3) * box_size
particle_positions += 0.05 * np.sin(5 * particle_positions)

# Synthetic vector field: simple 3D vortex-like field
x = y = z = np.linspace(0, box_size, N_grid)
X, Y, Z = np.meshgrid(x, y, z, indexing='ij')
VX = -np.sin(np.pi * Y) * np.cos(np.pi * Z)
VY =  np.sin(np.pi * Z) * np.cos(np.pi * X)
VZ = -np.sin(np.pi * X) * np.cos(np.pi * Y)
bulk_velocity = np.stack([VX, VY, VZ], axis=-1)

# Flatten for PyVista
points = np.vstack([X.ravel(), Y.ravel(), Z.ravel()]).T
vectors = bulk_velocity.reshape(-1, 3)

# Create structured grid
grid = pv.StructuredGrid()
grid.points = points
grid.dimensions = [N_grid, N_grid, N_grid]
grid['vectors'] = vectors

# Generate streamlines
streamlines = grid.streamlines('vectors', integration_direction='both',
                               initial_step_length=0.01, terminal_speed=1e-3)

# Plotting
plotter = pv.Plotter(off_screen=True, window_size=(1000, 1000))
plotter.add_mesh(streamlines.tube(radius=0.002), color='orange', opacity=0.4)
plotter.add_points(particle_positions, color='midnightblue', point_size=3.0, render_points_as_spheres=True)
plotter.set_background("white")
plotter.show_bounds(grid='front', location='outer', all_edges=True)

# Save output
plotter.screenshot("cosmology_streamlines_particles.png")
print("Saved to cosmology_streamlines_particles.png")
```

```python
import numpy as np

import yt

# Load the dataset.  We'll work with a some Gadget data to illustrate all
# the different ways in which the effective resolution can vary.  Specifically,
# we'll use the GadgetDiskGalaxy dataset available at
#  http://yt-project.org/data/GadgetDiskGalaxy.tar.gz

# load the data with a refinement criteria of 2 particle per cell
# n.b. -- in yt-4.0, n_ref no longer exists as the data is no longer
# deposited only a grid.  At present (03/15/2019), there is no way to
# handle non-gas data in Gadget snapshots, though that is work in progress
if int(yt.__version__[0]) < 4:
    # increasing n_ref will result in a "lower resolution" (but faster) image,
    # while decreasing it will go the opposite way

    ds = yt.load("./gevolution-1.3/output/snap001_cdm", n_ref=16)
else:
    ds = yt.load("./gevolution-1.3/output/snap001_cdm")

# Create projections of the density (max value in each resolution element in the image):
prj = yt.ProjectionPlot(
    ds, "x", ("gas", "density"), method="max", center="max", width=(100, "kpc")
)

# nicen up the plot by using a better interpolation:
plot = prj.plots[list(prj.plots)[0]]
ax = plot.axes
img = ax.images[0]
img.set_interpolation("bicubic")

# nicen up the plot by setting the background color to the minimum of the colorbar
prj.set_background_color(("gas", "density"))

# vary the buff_size -- the number of resolution elements in the actual visualization
# set it to 2000x2000
buff_size = 2000
prj.set_buff_size(buff_size)

# set the figure size in inches
figure_size = 10
prj.set_figure_size(figure_size)

# if the image does not fill the plot (as is default, since the axes and
# colorbar contribute as well), then figuring out the proper dpi for a given
# buff_size and figure_size is non-trivial -- it requires finding the bbox
# for the actual image:
bounding_box = ax.get_position()
# we're going to scale to the larger of the two sides
image_size = figure_size * max([bounding_box.width, bounding_box.height])
# now save with a dpi that's scaled to the buff_size:
dpi = np.rint(np.ceil(buff_size / image_size))
prj.save("with_axes_colorbar.png", mpl_kwargs=dict(dpi=dpi))

# in the case where the image fills the entire plot (i.e. if the axes and colorbar
# are turned off), it's trivial to figure out the correct dpi from the buff_size and
# figure_size (or vice versa):

# hide the colorbar:
prj.hide_colorbar()

# hide the axes, while still keeping the background color correct:
prj.hide_axes(draw_frame=True)

# save with a dpi that makes sense:
dpi = np.rint(np.ceil(buff_size / figure_size))
# prj.save("no_axes_colorbar.png", mpl_kwargs=dict(dpi=dpi))
```

```python
import yt
import numpy as np

ds = yt.load("./gevolution-1.3/output/snap001_cdm")

# Create a uniform grid for CIC density
ad = ds.all_data()

# Create deposited CIC field explicitly:
yt.add_particle_filter("all", function=None, filtered_type='all', requires=[])
ds.add_particle_filter('all')

dims = [64, 64, 64]
cic_density = ds.deposit('all', 'cic', dims=dims)

# Wrap it into a grid structure manually
left_edge = ds.domain_left_edge
right_edge = ds.domain_right_edge

cg = ds.covering_grid(level=0, left_edge=left_edge, dims=dims)

# Fill the CIC field into the covering grid manually
cg['deposit', 'all_cic'] = cic_density

# Now plot using SlicePlot
slc = yt.SlicePlot(
    ds,
    axis='z',
    fields=('deposit', 'all_cic'),
    center='c',
    width=(300_000, 'kpc/h'),
)

slc.set_log(('deposit', 'all_cic'), True)
slc.set_cmap(('deposit', 'all_cic'), 'inferno')
slc.save('cic_density_slice')
```
