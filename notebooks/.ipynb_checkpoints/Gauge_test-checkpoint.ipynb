```python
hiclass_sync = {
    "Omega_smg": -1,
    "Omega_Lambda": 0,
    "Omega_fld": 0,
    "h": 0.675560,
    "Omega_b": 4.827542e-02,
    "Omega_cdm": 2.637707e-01,
    "Omega_g": 5.418479e-05,
    "Omega_ur": 3.748335e-05,
    "A_s": 2.215000e-09,
    "n_s": 0.961900,
    "N_ncdm": 0,
    "z_max_pk": 100,
    "gravity_model": "propto_omega",
    # "non linear": "Halofit",
    "parameters_smg": f"3.000000e+06, 0, 0, 0, 1",
    "expansion_model": "w0wa",
    "expansion_smg": f" 6.878623e-01,-9.000000e-01, 0",
    "output_background_smg": 4,
    "extra metric transfer functions": "yes",
    "output": "dTk,vTk,mPk",
    "P_k_max_1/Mpc": 192,
}


M_hiclass_sync = Class()
M_hiclass_sync.set(hiclass_sync)
M_hiclass_sync.compute()


h = M_hiclass_sync.h()

k_pivot = 0.05
h = M_hiclass_sync.h()
As = hiclass_sync["A_s"]
ns = hiclass_sync["n_s"]


#############################################################################################
#                                                             Background

backgrounds = M_hiclass_sync.get_background()

rho_b = interp1d(backgrounds["z"], backgrounds["(.)rho_b"], kind="cubic")
rho_cdm = interp1d(backgrounds["z"], backgrounds["(.)rho_cdm"], kind="cubic")
rho_g = interp1d(backgrounds["z"], backgrounds["(.)rho_g"], kind="cubic")
rho_ur = interp1d(backgrounds["z"], backgrounds["(.)rho_ur"], kind="cubic")
rho_tot = interp1d(backgrounds["z"], backgrounds["(.)rho_tot"], kind="cubic")
p_tot = interp1d(backgrounds["z"], backgrounds["(.)p_tot"], kind="cubic")
rho_crit = interp1d(backgrounds["z"], backgrounds["(.)rho_crit"], kind="cubic")

H = interp1d(backgrounds["z"], backgrounds["H [1/Mpc]"], kind="cubic")
H_prime = interp1d(backgrounds["z"], backgrounds["H_prime"], kind="cubic")
H_prime_prime = interp1d(backgrounds["z"], backgrounds["H_prime_prime"], kind="cubic")

rho_smg = interp1d(backgrounds["z"], backgrounds["(.)rho_smg"], kind="cubic")
p_smg = interp1d(backgrounds["z"], backgrounds["(.)p_smg"], kind="cubic")

rho_smg_prime = interp1d(
    backgrounds["z"], backgrounds["(.)rho_smg_prime"], kind="cubic"
)
p_smg_prime = interp1d(backgrounds["z"], backgrounds["(.)p_smg_prime"], kind="cubic")

alpha_K = interp1d(backgrounds["z"], backgrounds["kineticity_smg"], kind="cubic")
alpha_B = interp1d(backgrounds["z"], backgrounds["braiding_smg"], kind="cubic")

alpha_K_prime = interp1d(
    backgrounds["z"], backgrounds["kineticity_prime_smg"], kind="cubic"
)
alpha_B_prime = interp1d(
    backgrounds["z"], backgrounds["braiding_prime_smg"], kind="cubic"
)

a_hi = 1.0 / (1.0 + backgrounds["z"])
a_hiclass = interp1d(backgrounds["z"], a_hi, kind="cubic")
z_hi = backgrounds["z"]

Hconf_hi = interp1d(backgrounds["z"], backgrounds["H [1/Mpc]"] * a_hi)
Hconf_prime_hi = interp1d(
    backgrounds["z"],
    (a_hi**2) * (backgrounds["H [1/Mpc]"] ** 2) + a_hi * backgrounds["H_prime"],
)

H0 = Hconf_hi(0)
Mpl2 = 1 / (3.0 * H0**2)

Normalization = As * (kk / k_pivot) ** (ns - 1.0)

def a(x):
    return 1 / (1 + x)


####################################################################################################3

##################################################
#                                                #
#        for pi and zeta calculation             #
#                                                #
##################################################

pk_pi_hi = [None] * len(z)
pk_pi_prime_hi = [None] * len(z)
pk_zeta_hi = [None] * len(z)




Pk_sync = [None] * len(z)
Pk_newt = [None] * len(z)

for zind, z_val in enumerate(z):

    Omega_mg = rho_smg(z_val) / rho_crit(z_val)
    Omega_rad = rho_g(z_val) / rho_crit(z_val)
    Omega_m = (rho_cdm(z_val) + rho_b(z_val)) / rho_crit(z_val)
    Omega_ur = rho_ur(z_val) / rho_crit(z_val)

    Transfers_hiclass_sync = M_hiclass_sync.get_transfer(z_val)

    kk_hi = Transfers_hiclass_sync["k (h/Mpc)"]

    #############
    delta_b_hiclass_sync = interp1d(
        Transfers_hiclass_sync["k (h/Mpc)"], Transfers_hiclass_sync["d_b"], kind="cubic"
    )
    theta_b_hiclass_sync = interp1d(
        Transfers_hiclass_sync["k (h/Mpc)"], Transfers_hiclass_sync["t_b"], kind="cubic"
    )
    #############
    #                         theta_cdm is zero in synchronous gauge
    delta_cdm_hiclass_sync = interp1d(
        Transfers_hiclass_sync["k (h/Mpc)"], Transfers_hiclass_sync["d_cdm"], kind="cubic"
    )
    #############
    delta_g_hiclass_sync = interp1d(
        Transfers_hiclass_sync["k (h/Mpc)"], Transfers_hiclass_sync["d_g"], kind="cubic"
    )
    theta_g_hiclass_sync = interp1d(
        Transfers_hiclass_sync["k (h/Mpc)"], Transfers_hiclass_sync["t_g"], kind="cubic"
    )
    #############
    delta_ur_hiclass_sync = interp1d(
        Transfers_hiclass_sync["k (h/Mpc)"], Transfers_hiclass_sync["d_ur"], kind="cubic"
    )
    theta_ur_hiclass_sync = interp1d(
        Transfers_hiclass_sync["k (h/Mpc)"], Transfers_hiclass_sync["t_ur"], kind="cubic"
    )
    #############

    delta_tot_hiclass_sync = interp1d(
        Transfers_hiclass_sync["k (h/Mpc)"], Transfers_hiclass_sync["d_tot"], kind="cubic"
    )

    theta_tot_hiclass_sync = interp1d(
        Transfers_hiclass_sync["k (h/Mpc)"], Transfers_hiclass_sync["t_tot"], kind="cubic"
    )

    phi_hiclass_sync = interp1d(Transfers_hiclass_sync["k (h/Mpc)"], Transfers_hiclass_sync["phi"], kind="cubic")

    phi_prime_hiclass_sync = interp1d(
        Transfers_hiclass_sync["k (h/Mpc)"], Transfers_hiclass_sync["phi_prime"], kind="cubic"
    )
    psi_hiclass_sync= interp1d(Transfers_hiclass_sync["k (h/Mpc)"], Transfers_hiclass_sync["psi"], kind="cubic")

    vx_smg = interp1d(Transfers_hiclass_sync["k (h/Mpc)"], Transfers_hiclass_sync["vx_smg"], kind="cubic")

    vx_prime_smg = interp1d(
        Transfers_hiclass_sync["k (h/Mpc)"], Transfers_hiclass_sync["vx_prime_smg"], kind="cubic"
    )
    eta = interp1d(Transfers_hiclass_sync["k (h/Mpc)"], Transfers_hiclass_sync["eta"], kind="cubic")

    eta_prime = interp1d(
        Transfers_hiclass_sync["k (h/Mpc)"], Transfers_hiclass_sync["eta_prime"], kind="cubic"
    )

    h_met = interp1d(Transfers_hiclass_sync["k (h/Mpc)"], Transfers_hiclass_sync["h"], kind="cubic")

    h_prime_met = interp1d(
        Transfers_hiclass_sync["k (h/Mpc)"], Transfers_hiclass_sync["h_prime"], kind="cubic"
    )
    
    
    
    alpha = (h_prime_met(kk) + 6.0 * eta_prime(kk)) / (2.0 * kk * kk * h * h)
    pi_hi = -vx_smg(kk) + alpha

    alpha_prime = psi_hiclass_sync(kk) - Hconf_hi(z_val) * alpha
    pi_prime_hi = -vx_prime_smg(kk) + alpha_prime

    zeta_hi = pi_prime_hi + Hconf_hi(z_val) * pi_hi - psi_hiclass_sync(kk)

    pk_pi_hi[zind] = Normalization * (pi_hi**2)
    
    pk_pi_prime_hi[zind] = Normalization * (pi_prime_hi**2)

    pk_zeta_hi[zind] = Normalization * (zeta_hi**2)

    tmp_delta_sync = []
    tmp_delta_newt = []

    delta_cdm_hiclass_newt = []
    theta_cdm_hiclass_newt = []
    
    delta_b_hiclass_newt = []
    theta_b_hiclass_newt = []
    
    delta_g_hiclass_newt = []
    theta_g_hiclass_newt = []
    
    delta_ur_hiclass_newt = []
    theta_ur_hiclass_newt = []
    
    delta_tot_hiclass_newt = []
    theta_tot_hiclass_newt = []
    

    for k in kk:
        alpha = (h_prime_met(k) + 6.0 * eta_prime(k)) / (2.0 * k * k * h * h)
        alpha_prime = psi_hiclass_sync(k) - Hconf_hi(z_val) * alpha

        pi_newt = -vx_smg(k) + alpha
        pi_prime_newt = -vx_prime_smg(k) + alpha_prime
        zeta_newt = Hconf_hi(z_val) * pi_newt + pi_prime_newt - psi_hiclass_sync(k)
        

        delta_smg_newt = ((1 + w0_fld) / cs2_fld) * (
            -(3 * cs2_fld - 1) * Hconf_hi(z_val) * (-vx_smg(k)) + (-vx_prime_smg(k))
        ) - 3*(1 + w0_fld)*Hconf_hi(z_val)*alpha

#         delta_smg_newt = (
#             (Mpl2 / a(z_val) ** 2)
#             * (
#                 -alpha_B(z_val) * Hconf_hi(z_val) * ((k * h) ** 2) * pi_newt
#                 + 3 * alpha_B(z_val) * (Hconf_hi(z_val) ** 2) * psi_syn(k)
#                 - (3 * alpha_B(z_val) + alpha_K(z_val))
#                 * (Hconf_hi(z_val) ** 2)
#                 * zeta_newt
#                 + 3 * alpha_B(z_val) * Hconf_hi(z_val) * phi_prime_syn(k)
#             )
#             + (
#                 (Mpl2 * alpha_B(z_val) / (a(z_val) ** 2))
#                 * (Hconf_prime_hi(z_val) - Hconf_hi(z_val) ** 2)
#                 + ((rho_smg(z_val) + p_smg(z_val)) / H0**2)
#             )
#             * 3
#             * Hconf_hi(z_val)
#             * pi_newt
#         ) / (rho_smg(z_val) / H0**2)

        ###########################################################################
        #                                          To Newtonian
        
        delta_cdm_hiclass_newt.append(delta_cdm_hiclass_sync(k) - 3 * alpha * Hconf_hi(z_val))
        theta_cdm_hiclass_newt.append(alpha *(k*k*h*h))

        delta_b_hiclass_newt.append(delta_b_hiclass_sync(k) - 3 * alpha * Hconf_hi(z_val))
        theta_b_hiclass_newt.append(theta_b_hiclass_sync(k) + alpha *(k*k*h*h))

        delta_g_hiclass_newt.append(delta_g_hiclass_sync(k) - 3 * alpha * Hconf_hi(z_val) * (1 + 1 / 3))
        theta_g_hiclass_newt.append(theta_g_hiclass_sync(k) + alpha *(k*k*h*h))

        delta_ur_hiclass_newt.append(delta_ur_hiclass_sync(k) - 3 * alpha * Hconf_hi(z_val) * (1 + 1 / 3))
        theta_ur_hiclass_newt.append(theta_ur_hiclass_sync(k) + alpha *(k*k*h*h))

#         d_tot_hiclass = (
#             rho_cdm(z_val) * d_cdm_hiclass
#             + rho_b(z_val) * d_b_hiclass
#             + rho_g(z_val) * d_g_hiclass
#             + rho_ur(z_val) * d_ur_hiclass
#             + rho_smg(z_val) * -delta_smg_newt
#         ) / rho_tot(z_val)

#         delta_tot_hiclass.append(d_tot_hiclass)
        

        #####################################################
        #                           Farbod's transformation


        delta_tot_hiclass_newt.append(delta_tot_hiclass_sync(k)- 3* alpha* (
                        Omega_mg * (1 - 0.9)
                        + Omega_m
                        + Omega_rad * (1 + 1 / 3)
                        + Omega_ur * (1 + 1 / 3)
                    ) * Hconf_hi(z_val))
        
        theta_tot_hiclass_newt.append(theta_tot_hiclass_sync(k)+ alpha *(k*k*h*h))

        ############################################################
        #                         another way

#         d_tot_hiclass =  delta_tot_hiclass_sync(k) - alpha * 3 * Hconf_hi(z_val) * (
#             rho_tot(z_val) + p_tot(z_val)
#         ) / rho_tot(z_val)

#         delta_tot_hiclass_newt.append(d_tot_hiclass)

        ###########################################################################
        #                                              sync
        delta_sync = (
            rho_cdm(z_val) * delta_cdm_hiclass_sync(k) + rho_b(z_val) * delta_b_hiclass_sync(k)
        ) / (rho_cdm(z_val) + rho_b(z_val))
        tmp_delta_sync.append(
            (2 * (np.pi**2) / ((k) ** 3) * As * (k * h / k_pivot) ** (ns - 1.0))
            * delta_sync**2
        )

        ###### ###################################################################
        #                                              Newtonian
        delta_newt = delta_sync - 3 * alpha * Hconf_hi(z_val)
        tmp_delta_newt.append(
            (2 * (np.pi**2) / ((k) ** 3) * As * (k * h / k_pivot) ** (ns - 1.0))
            * delta_newt**2
        )

    Pk_sync[zind] = tmp_delta_sync
    Pk_newt[zind] = tmp_delta_newt
```
