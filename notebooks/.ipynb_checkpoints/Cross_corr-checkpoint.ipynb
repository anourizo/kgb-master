```python
import os
import numpy as np

# ──────────────────────────────────────────────────────────────────────────────
# 1) PARAMETERS FOR YOUR SIMULATIONS

sim1 = dict(path="/home/ahmad/kgb-master/data/Cross-correlation/ak3e3-ab0/",    ngrid=1024, boxsize=400)
sim2 = dict(path="/home/ahmad/kgb-master/data/Cross-correlation/ak3e3-ab0.75/", ngrid=1024, boxsize=400)
sim3 = dict(path="/home/ahmad/kgb-master/data/Cross-correlation/ak3e3-ab1.5/",  ngrid=1024, boxsize=400)
sim4 = dict(path="/home/ahmad/kgb-master/data/Cross-correlation/ak3e6-ab0/",    ngrid=1024, boxsize=400)
sim5 = dict(path="/home/ahmad/kgb-master/data/Cross-correlation/ak3e6-ab0.75/", ngrid=1024, boxsize=400)
sim6 = dict(path="/home/ahmad/kgb-master/data/Cross-correlation/ak3e6-ab1.5/",  ngrid=1024, boxsize=400)
sim7 = dict(path="/home/ahmad/kgb-master/data/Cross-correlation/ak3e8-ab0/",    ngrid=1024, boxsize=400)
sim8 = dict(path="/home/ahmad/kgb-master/data/Cross-correlation/ak3e8-ab0.75/", ngrid=1024, boxsize=400)
sim9 = dict(path="/home/ahmad/kgb-master/data/Cross-correlation/ak3e8-ab1.5/",  ngrid=1024, boxsize=400)



# Put all the simulations in one dict. Add or remove entries here freely.
sims = {
    "sim1": sim1,
    "sim2": sim2,
    "sim3": sim3,
    "sim4": sim4,
    "sim5": sim5,
    "sim6": sim6,
    "sim7": sim7,
    "sim8": sim8,
    "sim9": sim9,
}

# Common list of P(k) suffixes
pknames = [
    "delta_kgb",
    "delta",
    "deltakgb_deltam",
]

# ──────────────────────────────────────────────────────────────────────────────
# 2) HELPERS

def read_redshifts(path_to_sim_ini: str):
    """Read redshifts from file_settings_used.ini in the given sim folder."""
    ini_path = os.path.join(path_to_sim_ini, "file_settings_used.ini")
    z = []
    with open(ini_path) as f:
        for line in f:
            if line.strip().startswith("Pk redshifts"):
                z = [float(x) for x in line.split("=")[1].split("#")[0].split(",")]
                break
    return z

def load_simulation(path: str, z, pknames):
    """Load k and P(k) arrays for all snapshots given by indices of z."""
    data = {name: [] for name in ["k"] + pknames}
    for i in range(len(z)):
        idx = f"{i:02d}"
        got_k = False
        for name in pknames:
            arr = np.loadtxt(os.path.join(path, f"pk_0{idx}_{name}.dat"))
            if not got_k:
                data["k"].append(arr[:, 0])
                got_k = True
            data[name].append(arr[:, 1])
    return data

def trim_snapshot(k, fields, ngrid, boxsize, remove_first=0, nyquist_frac=1):
    """
    Trim modes: drop the first 'remove_first' bins and cut at a fraction of the Nyquist.
    Uses np.searchsorted so it behaves correctly even if no k > cutoff.
    """
    kny = np.pi * ngrid / boxsize
    cutoff = nyquist_frac * kny
    # index of first k > cutoff; if none, iend = len(k)
    i_gt = np.where(k > cutoff)[0]
    iend = int(i_gt[0]) if i_gt.size > 0 else len(k)
    keep = slice(remove_first, iend)
    k_trimmed = k[keep]
    return k_trimmed, {name: arr[keep] for name, arr in fields.items()}

def process_sim(sim_cfg, z, pknames, remove_first=0, nyquist_frac=1/2):
    """Load one simulation and produce the trimmed output dict."""
    data = load_simulation(sim_cfg["path"], z, pknames)
    final = {name: [] for name in ["k"] + pknames}
    for iz in range(len(z)):
        k_acc, f_acc = trim_snapshot(
            data["k"][iz],
            {n: data[n][iz] for n in pknames},
            sim_cfg["ngrid"],
            sim_cfg["boxsize"],
            remove_first=remove_first,
            nyquist_frac=nyquist_frac,
        )
        final["k"].append(k_acc)
        for name in pknames:
            final[name].append(f_acc[name])
    return final

def sci_latex(x, sigfigs=2, trim_one=True):
    """
    Format x as LaTeX scientific notation:
      10^{-4}  or  3.2\\times 10^{-4}
    Handles the case where rounding makes the mantissa == 10.
    """
    if not np.isfinite(x) or x == 0:
        return "0"
    sign = "-" if x < 0 else ""
    x = abs(float(x))

    exp = int(np.floor(np.log10(x)))
    mant = x / (10 ** exp)

    # round mantissa to requested significant figures
    mant_rounded = float(f"{mant:.{sigfigs}g}")

    # if rounding pushes mantissa to 10, renormalize to 1 × 10^{exp+1}
    if mant_rounded >= 10.0:
        mant_rounded = 1.0
        exp += 1

    # if mantissa is ~1 and you prefer pure power-of-ten
    if trim_one and np.isclose(mant_rounded, 1.0, rtol=1e-12, atol=0.0):
        return rf"{sign}10^{{{exp}}}"

    mant_str = f"{mant_rounded:.{sigfigs}g}"
    return rf"{sign}{mant_str}\times 10^{{{exp}}}"


def make_cs2_interps(sims,
                     fname="file_background.dat",
                     z_col=3, cs2_col=21,
                     kind="cubic",
                     allow_extrapolation=True):
    """
    Returns:
      interps: dict sim_key -> callable f(z)
      series:  dict sim_key -> (z_sorted, cs2_sorted) raw arrays
    """
    interps = {}
    series = {}
    for key, cfg in sims.items():
        path = os.path.join(cfg["path"], fname)
        # read only the needed columns
        arr = np.loadtxt(path, usecols=(z_col, cs2_col))
        z, cs2 = arr[:, 0], arr[:, 1]

        # sort by z in case the file is descending
        order = np.argsort(z)
        z, cs2 = z[order], cs2[order]

        # cubic needs >=4 unique points; fall back to linear if needed
        uniq = np.unique(z)
        knd = kind if uniq.size >= 4 else "linear"

        f = interp1d(
            z, cs2, kind=knd,
            bounds_error=False,
            fill_value=("extrapolate" if allow_extrapolation else np.nan)
        )
        interps[key] = f
        series[key] = (z, cs2)
    return interps, series


# build all 9 in one shot
cs2_interps, cs2_series = make_cs2_interps(sims)


# ──────────────────────────────────────────────────────────────────────────────
# 3) DO THE WORK FOR *ALL* SIMULATIONS

# Read redshifts once (from sim1's INI as in your original)
z = read_redshifts(sim1["path"])

# Process every simulation automatically
results = {label: process_sim(cfg, z, pknames) for label, cfg in sims.items()}

# Access like:
# final1 = results["sim1"]
# final2 = results["sim2"]
# final3 = results["sim3"]

# === Build per-sim lookup dicts ===
def make_lookup(sim_key):
    return (
        dict(zip(z, results[sim_key]["k"])),
        dict(zip(z, results[sim_key]["deltakgb_deltam"])),
        dict(zip(z, results[sim_key]["delta_kgb"])),
        dict(zip(z, results[sim_key]["delta"]))
    )

k_zp_1, deltkdm_1, dkgb_1, dm_1 = make_lookup("sim1")
k_zp_2, deltkdm_2, dkgb_2, dm_2 = make_lookup("sim2")
k_zp_3, deltkdm_3, dkgb_3, dm_3 = make_lookup("sim3")
k_zp_4, deltkdm_4, dkgb_4, dm_4 = make_lookup("sim4")
k_zp_5, deltkdm_5, dkgb_5, dm_5 = make_lookup("sim5")
k_zp_6, deltkdm_6, dkgb_6, dm_6 = make_lookup("sim6")
k_zp_7, deltkdm_7, dkgb_7, dm_7 = make_lookup("sim7")
k_zp_8, deltkdm_8, dkgb_8, dm_8 = make_lookup("sim8")
k_zp_9, deltkdm_9, dkgb_9, dm_9 = make_lookup("sim9")

def corr_ratio(delt_kdm, d_kgb, d_m):
    # elementwise: deltakgb_deltam / sqrt(delta_kgb * delta_m)
    a = np.asarray(delt_kdm)
    b = np.asarray(d_kgb)
    c = np.asarray(d_m)
    denom = np.sqrt(b * c)
    with np.errstate(divide='ignore', invalid='ignore'):
        out = a / denom
    return out
```

```python
# ──────────────────────────────────────────────────────────────────────────────
# 0) IMPORTS & STYLE
import os
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import LogLocator
import seaborn as sns

plt.rc("text", usetex=True)
plt.rc("font", family="serif", size=25)
plt.rc("xtick", labelsize=20)
plt.rc("ytick", labelsize=20)

# ──────────────────────────────────────────────────────────────────────────────
# 1) SIM CONFIG

sim1 = dict(path="/home/ahmad/kgb-master/data/Cross-correlation/ak3e3-ab0/",    ngrid=1024, boxsize=400)
sim2 = dict(path="/home/ahmad/kgb-master/data/Cross-correlation/ak3e3-ab0.75/", ngrid=1024, boxsize=400)
sim3 = dict(path="/home/ahmad/kgb-master/data/Cross-correlation/ak3e3-ab1.5/",  ngrid=1024, boxsize=400)
sim4 = dict(path="/home/ahmad/kgb-master/data/Cross-correlation/ak3e6-ab0/",    ngrid=1024, boxsize=400)
sim5 = dict(path="/home/ahmad/kgb-master/data/Cross-correlation/ak3e6-ab0.75/", ngrid=1024, boxsize=400)
sim6 = dict(path="/home/ahmad/kgb-master/data/Cross-correlation/ak3e6-ab1.5/",  ngrid=1024, boxsize=400)
sim7 = dict(path="/home/ahmad/kgb-master/data/Cross-correlation/ak3e8-ab0/",    ngrid=1024, boxsize=400)
sim8 = dict(path="/home/ahmad/kgb-master/data/Cross-correlation/ak3e8-ab0.75/", ngrid=1024, boxsize=400)
sim9 = dict(path="/home/ahmad/kgb-master/data/Cross-correlation/ak3e8-ab1.5/",  ngrid=1024, boxsize=400)

sims = {
    "sim1": sim1, "sim2": sim2, "sim3": sim3,
    "sim4": sim4, "sim5": sim5, "sim6": sim6,
    "sim7": sim7, "sim8": sim8, "sim9": sim9,
}

# The fields we will read
pknames = ["delta_kgb", "delta", "deltakgb_deltam"]

# ──────────────────────────────────────────────────────────────────────────────
# 2) IO HELPERS

def read_redshifts(path_to_sim_ini: str):
    """Read redshifts from file_settings_used.ini in the given sim folder."""
    ini_path = os.path.join(path_to_sim_ini, "file_settings_used.ini")
    z = []
    with open(ini_path) as f:
        for line in f:
            if line.strip().startswith("Pk redshifts"):
                z = [float(x) for x in line.split("=")[1].split("#")[0].split(",")]
                break
    return z

def load_simulation(path: str, z, pknames):
    """Load k and P(k) arrays for all snapshots given by indices of z."""
    data = {name: [] for name in ["k"] + pknames}
    for i in range(len(z)):
        idx = f"{i:02d}"
        got_k = False
        for name in pknames:
            arr = np.loadtxt(os.path.join(path, f"pk_0{idx}_{name}.dat"))
            if not got_k:
                data["k"].append(arr[:, 0]); got_k = True
            data[name].append(arr[:, 1])
    return data

def trim_snapshot(k, fields, ngrid, boxsize, remove_first=0, nyquist_frac=1.0):
    """Drop the first 'remove_first' bins and cut at a fraction of the Nyquist."""
    kny = np.pi * ngrid / boxsize
    cutoff = nyquist_frac * kny
    i_gt = np.where(k > cutoff)[0]
    iend = int(i_gt[0]) if i_gt.size > 0 else len(k)
    keep = slice(remove_first, iend)
    k_trimmed = k[keep]
    return k_trimmed, {name: arr[keep] for name, arr in fields.items()}

def process_sim(sim_cfg, z, pknames, remove_first=0, nyquist_frac=1.0):
    """Load one simulation and produce trimmed arrays per snapshot."""
    data = load_simulation(sim_cfg["path"], z, pknames)
    final = {name: [] for name in ["k"] + pknames}
    for iz in range(len(z)):
        k_acc, f_acc = trim_snapshot(
            data["k"][iz],
            {n: data[n][iz] for n in pknames},
            sim_cfg["ngrid"], sim_cfg["boxsize"],
            remove_first=remove_first, nyquist_frac=nyquist_frac,
        )
        final["k"].append(k_acc)
        for name in pknames:
            final[name].append(f_acc[name])
    return final

# ──────────────────────────────────────────────────────────────────────────────
# 3) LOAD ALL SIMS

# Read redshifts once (from sim1's INI)
z = read_redshifts(sim1["path"])

# Optional trimming knobs for both IO and plotting
REMOVE_FIRST = 0
NYQUIST_FRAC = 1.0

results = {label: process_sim(cfg, z, pknames,
                              remove_first=REMOVE_FIRST,
                              nyquist_frac=NYQUIST_FRAC)
           for label, cfg in sims.items()}

# ──────────────────────────────────────────────────────────────────────────────
# 4) METRIC

def corr_ratio(deltakgb_deltam, delta_kgb, delta_m):
    """Compute deltakgb_deltam / sqrt(delta_kgb * delta_m) elementwise."""
    a = np.asarray(deltakgb_deltam)
    b = np.asarray(delta_kgb)
    c = np.asarray(delta_m)
    denom = np.sqrt(b * c)
    with np.errstate(divide="ignore", invalid="ignore"):
        out = a / denom
    return out

# ──────────────────────────────────────────────────────────────────────────────
# 5) PLOT: columns = redshift, 3 rows of sims (1–3, 4–6, 7–9)

# Choose which redshifts to show (by default: all)
# You can restrict, e.g. target_z = [z[i] for i in [0,3,6]]
target_z = z

# If you want an extra left-cut for plotting (in addition to REMOVE_FIRST), set here
START_IND = 0

row_groups = [
    ("sim1–sim3", ["sim1", "sim2", "sim3"]),
    ("sim4–sim6", ["sim4", "sim5", "sim6"]),
    ("sim7–sim9", ["sim7", "sim8", "sim9"]),
]

nrows = len(row_groups)
ncols = len(target_z)

fig, axes = plt.subplots(
    nrows, ncols,
    figsize=(5.2 * max(ncols, 2), 4.3 * nrows + 1.0),
    sharey=True,
    gridspec_kw=dict(wspace=0.10, hspace=0.32),
)

# Ensure 2D axes array even for edge cases
axes = np.atleast_2d(axes)

# Map redshift value -> snapshot index (values from the same list, so exact keys work)
z_to_idx = {zv: i for i, zv in enumerate(z)}

def format_axis(ax, z_val):
    ax.set_xscale("log")
    ax.set_yscale("log")
    ax.set_ylim(1e-2, 2e0)
    ax.yaxis.set_minor_locator(LogLocator(base=10.0, subs=np.arange(2, 10), numticks=12))
    ax.yaxis.set_major_locator(LogLocator(numticks=15))
    ax.grid(True, which="both", linestyle="--", alpha=0.15)
    ax.set_title(rf"$z={z_val}$", fontsize=22, pad=8)
    ax.set_xlabel(r"$k\,[h/\mathrm{Mpc}]$", fontsize=25)

for r, (row_name, sim_keys) in enumerate(row_groups):
    # consistent colors per simulation within this row (3 lines)
    colors = sns.color_palette("viridis", len(sim_keys)).as_hex()

    for c, z_val in enumerate(target_z):
        ax = axes[r, c]
        iz = z_to_idx[z_val]

        for color, key in zip(colors, sim_keys):
            simdat = results[key]
            k_vals = simdat["k"][iz]
            y_vals = corr_ratio(simdat["deltakgb_deltam"][iz],
                                simdat["delta_kgb"][iz],
                                simdat["delta"][iz])
            ax.plot(k_vals[START_IND:], y_vals[START_IND:],
                    "-", linewidth=2.1, color=color, label=f"KGB ({key})")

        if c == 0:
            ax.set_ylabel(r"$\Delta_{X}$", fontsize=25)

        format_axis(ax, z_val)

    # one legend per row (put it on the first column)
    handles, labels = axes[r, 0].get_legend_handles_labels()
    axes[r, 0].legend(handles, labels, loc="lower right", fontsize=16, bbox_to_anchor=(0.42, 0.00))

fig.tight_layout()
plt.show()
```
