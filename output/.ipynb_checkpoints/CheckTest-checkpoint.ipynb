```python
classtk_z100 = np.loadtxt("./HICLASS/file_classz1_tk.dat")
hiclass_background = np.loadtxt("./HICLASS/file_classbackground.dat")
```

```python
kk_hiclass= classtk_z100[0]
vx_smg = classtk_z100[5]
vx_prime_smg = classtk_z100[6]
z_hiclass = hiclass_background[:,0]
H_hiclass = hiclass_background[:,3]
```

```python
f_vx_smg = interp1d(kk_hiclass, vx_smg , kind="cubic")
f_vx_prime_smg = interp1d(kk_hiclass, vx_prime_smg, kind="cubic")
f_H_hi = interp1d(z_hiclass, H_hiclass, kind="cubic")
```

```python
kk, pi_pk100 = np.loadtxt("./pk_000_pi_k.dat",usecols=[0,1],unpack = True)
```

```python
Pk_pi100= []

for i in range (0,len(kk)):
    Pk_pi100.append(pi_pk100[i]*(2*(np.pi)**2)/((kk[i])**3))
```

```python
fig, ax = plt.subplots(1,1,figsize = (8,6),sharex=True)

plt.rc("text", usetex=True)
font = {"family": "normal", "weight": "bold", "size": 15}

ax.set_title('Ratio')   
ax.plot(kk, Pk_pi100)

ax.set_xscale('log')
ax.set_yscale('log')
ax.set_xlabel("$k~ [h/Mpc]$")
ax.set_ylabel(r"$\frac{P_{DE}~ [Mpc/h]^3}{P_{m}~ [Mpc/h]^3}$")

#ax.axvline(x = 5.02, c = 'red', linestyle = '--')

ax.grid()
```
